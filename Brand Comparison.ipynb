```python
lenovo=pd.read_csv("C:/Users/Lenovo/Downloads/Lenovo Laptop.csv" , index_col=0)
lenovo
```

```python
# Merging Lenovo and msi data as Laptopdata

df1=lenovo.join(msi)
Laptopdata=pd.DataFrame(df1)
Laptopdata
```

```python
Laptopdata.describe()
```

```python
Laptopdata.info()
```

```python
print(Laptopdata.isnull().sum(axis=0))
```

```python
plt.figure(figsize=(8,7))
Laptopdata.plot(x ='Product_name1', y=['Price1'], kind='bar')
plt.xlabel("Product Name")
plt.ylabel("Price")
plt.title("Comapring Price of LENOVO")
plt.legend("Price")
```

```python
plt.figure(figsize=(8,7))
Laptopdata.plot(x ='Product_name2', y=['Price2'], kind='bar')
plt.xlabel("Product Name")
plt.ylabel("Price")
plt.title("Comapring Price of MSI")
plt.legend("Price")
```

```python
plt.figure(figsize=(15,7))
r=plt.subplot()
r.bar(Laptopdata['Discount1'],Laptopdata['Rating1'])
r.bar(Laptopdata['Discount2'],Laptopdata['Rating2'])
plt.xlabel("Discount Price")
plt.ylabel("Rating")
plt.legend(['Lenovo','MSI'] )
```
