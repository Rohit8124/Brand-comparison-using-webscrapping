```python
# Importing libraries
from bs4 import BeautifulSoup as s
from urllib.request import urlopen as u
```

```python
# adding link 

link="https://www.flipkart.com/search?q=msi+laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_10_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_10_sc_na_na&as-pos=1&as-type=HISTORY&suggestionId=msi+laptop%7CLaptops&requestId=62c52c4b-0323-4077-83d5-f671259c2d05"
pagecontent=u(link)
pagecode=pagecontent.read()
```

```python
#read page
page_file=s(pagecode)
```

```python
pagecode
```

```python
name=page_file.find("div",{"class":'_4rR01T'})
print(name.text)
```

```python
productname=[]
name=page_file.findAll("div",{"class":'_4rR01T'})
for i in name:
    p=i.text
    productname.append(p)
```

```python
print(productname)
len (productname)
```

```python
# Rating

rating=[]
name=page_file.findAll("div",{"class":'_3LWZlK'})
for i in name:
    p=i.text
    rating.append(p)
```

```python
print(rating)
len(rating)
```

```python
#deleting Extra rating

del rating[-11:-1]
len(rating)
```

```python
# Price of the product

price=[]
name=page_file.findAll("div",{"class":'_30jeq3 _1_WHN1'})
for i in name:
    p=i.text
    price.append(int(p[1:].replace(',','')))
```

```python
print(price)
len(price)
```

```python
discount=[]
name=page_file.findAll("div",{"class":'_3Ay6Sb'})
for i in name:
    p=i.text
    discount.append(p)
```

```python
print(discount)
len(discount)
```

```python
del discount[-6:-1]
len(discount)
```

```python
# Importing library and creating dataframe

import pandas as pd
d={'Product_name2':productname,'Rating2':rating,'Price2':price,'Discount2':discount}
data=pd.DataFrame(d)
```

```python
data
```

```python
#Removing last row  so as to keep only 12 records

data.drop(data.tail(11).index,inplace=True)
data
```
